```python
import pandas as pd
import numpy as np
import nltk
import os
from nltk import word_tokenize
from nltk.probability import FreqDist
import urllib.request
from matplotlib import pyplot as plt3
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
nltk.download('punkt')
```

```python
def get_top_nwords(corpus):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq
```

```python
file = 'job_description.csv'
df = pd.read_csv(file)

# title = ['Marketing Manager']
# title = ['Director Business Analyst', 'Business Analyst', 'Senior Business Analyst']
# title = ['Product Manager', 'Senior Product Manager']
# title = ['Software Engineer', 'Senior Software Engineer']
title = ['Digital Marketing Specialist']
# title = ['Data Scientist']
# title = ['Data Analyst']
# title = ['Data Engineer']
# title = ['Business Intelligence Analyst']
# title = ['Financial Analyst']
df1 = df[df.title.isin(title)]


corpus = df1.description.str.cat(sep=' ')
corpus = corpus.lower()
tokenizer = RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(corpus)

# words = get_top_nwords(corpus)

# print(len(words))
```

```python
#create an empty list to store words
words_no_punc = []

#iterate through the words list to remove punctuations
for word in words:
    if word.isalpha():
        words_no_punc.append(word.lower())
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords

stopwords_list = stopwords.words("english")
```

```python
#create an empty list to store clean words
clean_words = []

#Iterate through the words_no_punc list and add non stopwords to the new clean_words list
for word in words:
    if word not in stopwords_list:
        clean_words.append(word)
```

```python
fdist = FreqDist(clean_words)


#print the 10 most common words
fdist.most_common(1000)
```
